```python
books.drop(['Image-URL-S', 'Image-URL-M', 'Image-URL-L'], axis=1,inplace=True)
```

```python
books.rename(columns={'Book-Title':'title','Book-Author':'author', 'Year-Of-Publication':'year', 'Publisher':'publisher'},inplace=True)
```

```python
ratings.rename(columns={'User-ID':'user_id','Book_Rating':'rating'},inplace=True)
```

```python
users.rename(columns={'User-ID':'user_id', 'Location':'location', 'Age':'age'},inplace=True)
```

```python
a=ratings['user_id'].value_counts()
```

```python
b=ratings['user_id'].value_counts()>200
```

```python
new_users=list(a[b].index)
```

```python
ratings=ratings[ratings['user_id'].isin(new_users)]
```

```python
ratings_with_bookname=ratings.merge(books, on='ISBN')
```

```python
ratings_with_bookname.shape
```

```python
ratings_with_bookname.head()
```

```python
book_num_ratings=ratings_with_bookname.groupby('title')['Book-Rating'].count().reset_index()
```

```python
book_num_ratings.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
```

```python
book_num_ratings.shape
```

```python
book_num_ratings=book_num_ratings[book_num_ratings['num_ratings']>=50]
```

```python
book_num_ratings.shape
```

```python
final_rating=ratings_with_bookname.merge(book_num_ratings, on='title')
```

```python
final_rating.shape
```

```python
final_rating=final_rating.drop_duplicates(subset=['user_id','title'])
```

```python
final_rating.shape
```

```python
final_rating.columns
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='Book-Rating')
```

```python
book_pivot.shape
```

```python
book_pivot=book_pivot.fillna(0)
```

```python
book_pivot
```

```python
#nearest neighbour algorithm
```

```python
from scipy.sparse import csr_matrix
book_matrix=csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_matrix)
```

```python
book_pivot.index[237]
```

```python
book_pivot.iloc[237,:].values.reshape(1,-1).shape
```

```python
distance,suggestion=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=5)
```

```python
suggestion
```

```python
book_pivot.index[184]
```

```python
for i in suggestion[0]:
    print(book_pivot.index[i])
```

```python
np.where(book_pivot.index=='Exclusive')[0][0]
```

```python
def recommend(bookname):
    b_id=np.where(book_pivot.index=='Exclusive')[0][0]
    distance,suggestion=model.kneighbors(book_pivot.iloc[b_id,:].values.reshape(1,-1),n_neighbors=5)
    for i in suggestion[0]:
        print(book_pivot.index[i])
```

```python
recommend('Harry Potter and the Chamber of Secrets ')
```
